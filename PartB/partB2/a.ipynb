```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../Data/Hotels_data_Changed.csv')
df.head()
```

```python
df.drop(["Days", "Original Price", "Discount Price", "Available Rooms", "Hotel Stars", "DiscountDiff"] , axis = 1, inplace = True)
df.head()
```

```python
df["Snapshot Date"] = pd.to_datetime(df["Snapshot Date"])
df["Checkin Date"] = pd.to_datetime(df["Checkin Date"])
df["Snapshot Date"] = (df["Snapshot Date"] - df["Snapshot Date"].min()).dt.days
df["Checkin Date"] = (df["Checkin Date"] - df["Checkin Date"].min()).dt.days
df.head()
```

```python
week_mapping = {
    "Monday": 1, "Tuesday": 2, "Wednesday": 3, "Thursday": 4,
    "Friday": 5, "Saturday": 6, "Sunday": 7
}
df["WeekDay"] = df["WeekDay"].map(week_mapping)
df.head()
```

```python
le_hotel = LabelEncoder()
df["Hotel_Encoded_Name"] = le_hotel.fit_transform(df["Hotel Name"])
df.head()
```

```python
df.drop(["Hotel Name"], axis = 1, inplace = True)   
df.head()
```

```python
df_best_discount = df.loc[df.groupby(["Snapshot Date", "Checkin Date", "DayDiff", "WeekDay", "Hotel_Encoded_Name"])["DiscountPerc"].idxmax()]
df_best_discount.head()
```

```python
df_best_discount.drop(["DiscountPerc"] , axis = 1, inplace = True)
df_best_discount.head()
```

```python
df_best_discount.to_csv("../Data/Hotels_Best_Discount.csv", index = False)
print("File saved successfully")
```
